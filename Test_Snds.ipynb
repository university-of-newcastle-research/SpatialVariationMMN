```python
from psychosnd import scla
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(rc={'figure.figsize':(11.7,6)})
```

```python
ptb_Primary = scla.scla(soundfile='timing_test/PTB_48k_Primary.wav', logfile='timing_test/PTB_Primary_test.csv', snddur=0.35, portdur=0.35, sndthresh=0.1, portthresh=0.1)
ptb_ASIO = scla.scla(soundfile='timing_test/PTB_48k_ASIO.wav', logfile='timing_test/PTB_ASIO_test.csv', snddur=0.35, portdur=0.35, sndthresh=0.1, portthresh=0.1)
ptb_Audigy = scla.scla(soundfile='timing_test/PTB_48k_Audigy.wav', logfile='timing_test/PTB_Audigy_test.csv', snddur=0.35, portdur=0.35, sndthresh=0.1, portthresh=0.1)
ptb_when = scla.scla(soundfile='timing_test/whentest.wav', logfile='timing_test/whentest.csv', snddur=0.35, portdur=0.35, sndthresh=0.1, portthresh=0.1)
```

```python
raw = {
    'ptb_primary': ptb_Primary['Port_to_Snd']['rawdata'],
    'ptb_ASIO': ptb_ASIO['Port_to_Snd']['rawdata'],
    'ptb_audigy': ptb_Audigy['Port_to_Snd']['rawdata'],
    'whentest': ptb_when['Port_to_Snd']['rawdata']
}
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in raw.items() ]))
df
```

```python
df.to_csv('soundtesting.csv')
```

```python
sns.kdeplot(data=df)
plt.show()
```

```python
sns.histplot(data=df, bins=100)
plt.show()
```
